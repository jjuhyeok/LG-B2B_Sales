```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd


data = pd.read_csv('/content/drive/MyDrive/LG_Aimers4/train.csv')
data2 = pd.read_csv('/content/drive/MyDrive/LG_Aimers4/submission.csv')
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

features = ['customer_type', 'customer_job', 'lead_desc_length', 'product_subcategory', 'product_model_name']

for feature in features:
    #data = pd.concat([data,data2],axis=0)
    unique_values = data[feature].fillna('Unknown').unique()
    embeddings = model.encode(unique_values.astype(str))
    similarity_matrix = cosine_similarity(embeddings)
    similarity_scores = np.mean(similarity_matrix, axis=1)
    value_score_mapping = dict(zip(unique_values, similarity_scores))
    data[f'{feature}_similarity_score'] = data[feature].map(value_score_mapping)
print(data.head())
```

그럴싸한 문장들 필터링

### 1. customer_type (고객 유형)
- **이유**: 고객 유형에 따라 제품에 대한 필요성이나 선호도가 다름
- **기대효과**: Ex)End-Customer와 유사한 유형의 표현 가능

### 2. customer_job (고객의 직업군)
- **이유**: 직업군은 구매력과 관심사에 영향을 미침
- **기대효과**: Ex)"IT 전문가"와 유사한 직업군을 표현 가능

### 3. expected_timeline (요청)
- **이유**: 설명 길이는 고객의 관심도와 요구사항의 상세함을 나타냅니다.
- **기대효과**: Ex) budget 들어있는 샘플들 뜯어보니까 0.08%

### 4. product_subcategory (요청 제품 하위 카테고리)
- **이유**: 제품 하위 카테고리는 고객의 구체적인 필요를 반영합니다.
- **기대효과**: 비슷한 하위 카테고리의 제품에 대한 고객의 선호도를 예측하여 타겟 마케팅에 활용할 수 있습니다.

### 5. product_model_name (제품 모델명)
- **이유**: 제품 모델명은 제품의 구체적인 특성과 관련이 있습니다.
- **기대효과**: 비슷한 모델명을 선호하는 고객 그룹을 식별하여, 그들의 선호도에 맞는 제품을 추천할 수 있습니다.

이러한 유사성 점수를 통해 각 피처별로 고객 데이터의 세밀한 패턴을 파악하고, 더 정확한 고객 세분화 및 타겟팅 전략을 수립할 수 있는 기대효과가 있습니다.
