```python
train = pd.read_csv("/content/drive/MyDrive/LG_Aimers4_본선/train.csv")
test = pd.read_csv("/content/drive/MyDrive/LG_Aimers4_본선/submission.csv")
```

```python
# 각 피처별 고유값의 수를 계산합니다.
nunique_values = train.nunique()

# 35개 이하의 고유값을 가진 피처들을 필터링합니다.
features_with_few_unique_values = nunique_values[nunique_values <= 35].index.tolist()

# 타겟 변수 'is_converted'를 제외한 피처 리스트를 생성합니다.
features_to_encode = [feature for feature in features_with_few_unique_values if feature != 'is_converted']

# 각 피처에 대해 타겟 인코딩을 적용하고 원본 데이터프레임에 추가합니다.
for feature in features_to_encode:
    # 각 카테고리별로 'is_converted'의 평균값을 계산합니다.
    encoded_feature_name = f"{feature}_encoded"
    train[encoded_feature_name] = train.groupby(feature)['is_converted'].transform('mean')
# For each feature with 35 or fewer unique values, calculate value_counts()
value_counts_per_feature = {feature: train[feature].value_counts() for feature in features_to_encode}
value_counts_per_feature
```

```python
# 중복 행을 제거하고 각 중복 그룹에서 첫 번째 행만 남김
train = train.drop_duplicates()
train = train.reset_index(drop = True)
train
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import nltk

# Downloading NLTK resources (stopwords, punkt for tokenization, wordnet for lemmatization)
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')

# Define a function for text preprocessing
def preprocess_text(text):
    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    # Remove non-letters and lower case all words
    text = re.sub(r'[^a-zA-Z]', ' ', text).lower()
    # Tokenize
    words = word_tokenize(text)
    # Remove stopwords
    words = [w for w in words if w not in stopwords.words('english')]
    # Lemmatize
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(w) for w in words]
    # Join the words back into one string
    return ' '.join(words)

# Applying the preprocessing function to the 'lead_description' column
train['lead_description_clean'] = train['lead_description'].astype(str).apply(preprocess_text)

# Vectorizing the preprocessed text
tfidf_vectorizer = TfidfVectorizer(max_features=1000)  # Limiting to 1000 features for demonstration
tfidf_matrix = tfidf_vectorizer.fit_transform(train['lead_description_clean'])

tfidf_matrix.shape
```
